```python
## Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge = Ridge()
```

```python
# defining parameters
params = {'alpha':[1e-15,1e-10,1e-5,1e-3,1e-2,1,2,5,10,20,30,40,45,50,60,70,80,90,100]}
ridge_regression = GridSearchCV(ridge,params,scoring = 'neg_mean_squared_error',cv = 5)
ridge_regression.fit(x_train,y_train)
```

```python
print(ridge_regression.best_params_,ridge_regression.best_score_)
```

```python
from sklearn.metrics import r2_score
y_pred = ridge_regression.predict(x_test)

r2_scores = r2_score(y_pred,y_test)
print(r2_scores)
```

```python
## Lasso Regression(L1)
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()

# defining parameters
params = {'alpha':[1e-15,1e-10,1e-5,1e-3,1e-2,1,2,5,10,20]}
lasso_regression = GridSearchCV(lasso,params,scoring = 'neg_mean_squared_error',cv = 10)
lasso_regression.fit(x_train,y_train)
```

```python
print(lasso_regression.best_params_)
print(lasso_regression.best_score_)
```

```python
from sklearn.metrics import r2_score
y_pred = lasso_regression.predict(x_test)

r2_scores = r2_score(y_pred,y_test)
print(r2_scores)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
breast_cancer
```

```python
x = pd.DataFrame(breast_cancer['data'],columns=breast_cancer['feature_names'])
```

```python
x
```

```python
# dataset['target']= breast_cancer['target']
y = pd.DataFrame(breast_cancer['target'],columns = ['target'])
```

```python
y.value_counts()
```

```python
logi_regression = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42) 
```

```python
params = {'C':[1,5,10],'max_iter':[100,150]}
```

```python
model = GridSearchCV(logi_regression,param_grid= params, scoring='f1',cv= 5)
```

```python
model.fit(x_train,y_train)
```

```python
print(model.best_params_,model.best_score_)
```

```python
y_pred= model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
